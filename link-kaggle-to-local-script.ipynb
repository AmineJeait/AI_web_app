````python
!kill $(lsof -t -i:18080) 2>/dev/null || echo "No process using port 18080"

from flask import Flask, request, jsonify
from pyngrok import ngrok
import threading
import time
import os
import pandas as pd
import matplotlib.pyplot as plt
import ollama

# ===== CONFIGURATION =====
PORT = 18080  # Try a different port if needed (e.g., 18081)
FRESH_NGROK_TOKEN = "306cpDOWm1Z6FyIlOZCzWQg9Wsi_5nx7Raifhdba1u1rs1WbJ"  # Your ngrok token
MODEL_NAME = "codellama:7b"  # Ollama model name
BENCHMARK_FILE = "gpu_benchmarks.csv"
# ========================

# Initialize Ngrok
ngrok.set_auth_token(FRESH_NGROK_TOKEN)

# Benchmark storage
if not os.path.exists(BENCHMARK_FILE):
    pd.DataFrame(columns=[
        'timestamp', 
        'device',
        'gpu_name',
        'model',
        'input_length',
        'output_length',
        'generation_time',
        'tokens_per_second'
    ]).to_csv(BENCHMARK_FILE, index=False)

def log_benchmark(data):
    """Save benchmark data to CSV"""
    df = pd.DataFrame([data])
    df.to_csv(BENCHMARK_FILE, mode='a', header=False, index=False)

# Initialize Ollama
try:
    print("Pulling Ollama model...")
    ollama.pull(MODEL_NAME)
    print(f"✅ Model {MODEL_NAME} pulled successfully!")
except Exception as e:
    print(f"❌ Ollama setup failed: {str(e)}")
    print("\nTroubleshooting Steps:")
    print("1. Make sure you have GPU enabled in Notebook Settings")
    print("2. Ensure internet access is turned ON")
    print("3. Try runtime restart")
    raise

app = Flask(__name__)

def generate_code_prompt(user_input):
    """Optimized prompt for complete code generation"""
    return f"""You are an expert Python programmer. Write COMPLETE, executable Python code for:
{user_input}

Requirements:
1. Include ALL necessary imports
2. Add all required functions/classes
3. Include complete implementation (don't omit parts)
4. Add error handling where appropriate
5. Return ONLY the code block in markdown format

Response format:
```python
# Complete code here
```"""

@app.route('/predict', methods=['POST'])
def predict():
    try:
        data = request.get_json()
        user_input = data.get('text', '').strip()
        
        if not user_input:
            return jsonify({'error': 'Empty input'}), 400
        
        # Use a more specific prompt for complete games
        prompt = f"""You are an expert game developer. Write a COMPLETE, playable Python implementation of {user_input} using Pygame.

Requirements:
1. Include ALL necessary imports
2. Implement all game objects (player, enemies, items)
3. Include collision detection
4. Add score tracking
5. Implement game loop with event handling
6. Make sure the code can run directly when copied
7. Return ONLY the complete code block wrapped in ```python``` markers

Complete code:
```python
"""
        
        response = ollama.generate(
            model=MODEL_NAME,
            prompt=prompt,
            options={
                'temperature': 0.2,  # Lower for more deterministic output
                'top_p': 0.9,
                'num_predict': 4096  # Increased for larger games
            }
        )
        
        full_response = response['response']
        
        # Extract the complete code block
        if '```python' in full_response:
            code_block = full_response.split('```python')[1].split('```')[0].strip()
        else:
            code_block = full_response.strip()
        
        return jsonify({
            'code': code_block,
            'warnings': []
        })
        
    except Exception as e:
        return jsonify({'error': str(e)}), 500

@app.route('/benchmarks', methods=['GET'])
def get_benchmarks():
    """Endpoint to retrieve benchmark data"""
    try:
        df = pd.read_csv(BENCHMARK_FILE)
        return jsonify({
            'benchmarks': df.to_dict(orient='records'),
            'stats': {
                'average_tokens_per_second': df['tokens_per_second'].mean(),
                'total_generations': len(df),
                'devices_tested': df['gpu_name'].unique().tolist()
            }
        })
    except Exception as e:
        return jsonify({'error': str(e)}), 500

def visualize_benchmarks():
    """Generate performance visualization"""
    try:
        df = pd.read_csv(BENCHMARK_FILE)
        if len(df) == 0:
            return
        
        plt.figure(figsize=(12, 6))
        
        # Plot tokens per second by device
        plt.subplot(1, 2, 1)
        df.groupby('gpu_name')['tokens_per_second'].mean().plot(
            kind='bar', 
            title='Average Tokens/Second',
            ylabel='Tokens/Second',
            color='skyblue'
        )
        
        # Plot generation time distribution
        plt.subplot(1, 2, 2)
        for device in df['gpu_name'].unique():
            device_data = df[df['gpu_name'] == device]
            plt.scatter(
                device_data['output_length'],
                device_data['generation_time'],
                label=device,
                alpha=0.6
            )
        plt.title('Generation Time vs Output Length')
        plt.xlabel('Output Tokens')
        plt.ylabel('Time (seconds)')
        plt.legend()
        
        plt.tight_layout()
        plt.savefig('benchmark_results.png')
        plt.close()
        
        print("✅ Benchmark visualization saved as benchmark_results.png")
    except Exception as e:
        print(f"❌ Failed to generate visualization: {str(e)}")

def run_flask():
    print(f"\nStarting Flask on port {PORT}...")
    app.run(host='0.0.0.0', port=PORT, threaded=True)

# Start server on a different port if needed
try:
    flask_thread = threading.Thread(target=lambda: app.run(host='0.0.0.0', port=PORT, threaded=True), daemon=True)
    flask_thread.start()
    time.sleep(3)
    
    # Start Ngrok
    public_url = ngrok.connect(PORT)
    print(f"\n✅ Ollama Code Generator Ready! Test with:")
    print(f'curl -X POST {public_url}/predict -H "Content-Type: application/json" -d \'{{"text":"Write a complete Python game"}}\'')
    
except Exception as e:
    print(f"\n❌ Error starting server: {str(e)}")
    print("Try changing the PORT number if issues persist")

# Add cell to display benchmark results
from IPython.display import Image, display

def show_benchmarks():
    try:
        df = pd.read_csv(BENCHMARK_FILE)
        if len(df) > 0:
            display(df.tail(5))
            
            if os.path.exists('benchmark_results.png'):
                display(Image(filename='benchmark_results.png'))
            else:
                print("Run predictions first to generate benchmark data")
        else:
            print("No benchmark data yet. Make some predictions first.")
    except Exception as e:
        print(f"Error displaying benchmarks: {str(e)}")

print("\nUse show_benchmarks() to view performance data")
````
